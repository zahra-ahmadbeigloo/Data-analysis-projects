# Plotting Map using Folium 

```python
import numpy as np
import pandas as pd
import folium
from folium import plugins
```

```python
world_map = folium.Map(tiles='Cartodb positron')
world_map
```

```python
df_crime = pd.read_csv('https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/Police_Department_Incidents_-_Previous_Year__2016_.csv')
df_crime.head()
```

```python
print(df_crime.shape)
df_crime = df_crime.iloc[0:200, :]
```

```python
sanfran_map = folium.Map(location=[37.77, -122.42], zoom_start=12)

crimes = folium.map.FeatureGroup()
for lat, lng, in zip(df_crime.Y, df_crime.X):
    crimes.add_child(
        folium.vector_layers.CircleMarker(
            [lat, lng], radius=5, color='red', fill=True, fill_color='white', fill_opacity=0.8))

sanfran_map.add_child(crimes)
```

```python
sanfran_map = folium.Map(location=[37.77, -122.42], zoom_start=12)

crimes = folium.map.FeatureGroup()

for lat, lng, in zip(df_crime.Y, df_crime.X):
    crimes.add_child(folium.vector_layers.CircleMarker(
    [lat, lng], radius=5, color='yellow', fill_color='blue', fill_opacity=0.6))

latitudes = list(df_crime.Y)
longitudes = list(df_crime.X)
labels = list(df_crime.Category)

for lat, lng, label in zip(latitudes, longitudes, labels):
    folium.Marker([lat, lng], icon=folium.Icon(icon='info_icon', color='lightblue'), popup=label).add_to(sanfran_map)    
    
sanfran_map.add_child(crimes)
```

```python
sanfran_map = folium.Map(location=[37.77, -122.42], zoom_start=12)

for lat, lng, label in zip(df_crime.Y, df_crime.X, df_crime.Category):
    folium.vector_layers.CircleMarker(
    [lat, lng], popup=label, radius=5, color='yellow', fill_color='blue', fill_opacity=0.6).add_to(sanfran_map)
   
sanfran_map
```

```python
sanfran_map = folium.Map(location = [37.77, -122.42], zoom_start = 12)

c = plugins.MarkerCluster().add_to(sanfran_map)

for lat, lng, label, in zip(df_crime.Y, df_crime.X, df_crime.Category):
    folium.Marker(location=[lat, lng], icon=None, popup=label).add_to(c)

sanfran_map
```

```python
df_can = pd.read_csv('https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/Canada.csv')
df_can.head()
```

```python
world_geo = r'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/world_countries.json'

world_map = folium.Map(location=[0, 0], zoom_start=2)
folium.Choropleth(
    geo_data=world_geo,
    data=df_can,
    columns=['Country', 'Total'],
    key_on='feature.properties.name',
    fill_color='YlOrRd', 
    fill_opacity=0.7, 
    legend_name='Immigration to Canada'
).add_to(world_map)

world_map
```
